# Querying a DataFrame

```python
import pandas as pd
```

```python
df = pd.read_csv('datasets/admissionpredict.csv', index_col = 0)
df.columns = [x.lower().strip() for x in df.columns]
df.head()
```

```python
# boolean masks are created by applying operators directly to our pd series or dataframa

# example - candidtes with chance greater than 0.7
df_mask = df['chance of admit'] > 0.7
df_mask # this returns the boolean mask for the given condition
```

```python
# now, we can lay this boolean mask on top of our dataframe
df.where(df_mask).head()
```

```python
# we can drop the NaN rows, which signify the values that dont match our boolean mask
df.where(df_mask).dropna().head()
```

```python
# a more efficient way to do this, that combines where() and dropna():
df[df['chance of admit'] > 0.7] # overload the indexing operator and directly sending the boolean mask in as a parameter
```

```python
# combining multiple boolean masks (& - and, | - or)
mask1 = (df['chance of admit'] > 0.7) & (df['chance of admit'] < 0.9)
df[mask1]
```

```python
# we can also use the .gt()(greater than) and the .lt()(lesser than) built in functions
mask2 = (df['chance of admit'].gt(0.7)) & (df['chance of admit'].lt(0.9))
df[mask2]
```
