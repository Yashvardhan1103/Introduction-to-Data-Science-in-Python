# Indexing DataFrames

```python
# we have seen that indexes can either be explicitly set (dictionaries) or they can be set implicitly (integers). we can also use the setindex() function
```

```python
df = pd.read_csv('datasets/admissionpredict.csv', index_col = 0)
df.head()
```

```python
# now, if we use set_index() to change the index column, the current index column will be lost. in order to avoid that:
df['Serial No.'] = df.index
df = df.set_index('Chance of Admit ')
df.head()
```

```python
# to remove index entirely:
df = df.reset_index()
df.head()
```

```python
# we can also set multiple columns as the index (multilevel indexing)
df = pd.read_csv('datasets/census.csv')
df.head()
```

```python
# in order to see the unique summarized levels:
df['SUMLEV'].unique()
# we can see that all elements of the dataset have sumlev either 40 or 50
```

```python
# now, we will see the rows that have sumlev = 50
df = df[df['SUMLEV'] == 50]
df.head()
```

```python
# now, we will just take a look at a few columns (births and population estimate for 2010 and 2011)
columns_to_keep = ['STNAME', 'CTYNAME', 'BIRTHS2010', 'BIRTHS2011', 'POPESTIMATE2010', 'POPESTIMATE2011']
df = df[columns_to_keep]
df.head()
```

```python
# we will set the state and county names as index
df = df.set_index(['STNAME', 'CTYNAME'])
df.head()
```

```python
# to query a dataframe with multiple indexes using .loc() [example - find these values for Dearborn County, Indiana]
df.loc['Indiana', 'Dearborn County']
```

```python
# we can also view multiple different county-state pairs by passing a list of tuples
df.loc[[('Indiana', 'Dearborn County'), 
       ('Iowa', 'Dickinson County'),
       ('Michigan', 'Wayne County')]]
```
